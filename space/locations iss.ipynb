```python
start = 1504843200
end = 1536426704
interval = 17520

lat = []
lon = []
vel = []
alt= []
ts = []


y = 0
for x in range (start, end, interval):
    url = f"https://api.wheretheiss.at/v1/satellites/25544/positions?timestamps={x}&units=miles"
    print(f'{x}')
    response = requests.get(url)
    location_json = response.json()
    # Add code to digest JSON and save to a list somewhere
    lat.append(location_json[0]["latitude"])
    lon.append(location_json[0]["longitude"])
    vel.append(location_json[0]["velocity"])
    alt.append(location_json[0]["altitude"])
    ts.append(location_json[0]["timestamp"])
    #this orrr
    time.sleep(2)
    
    #this
#    y = y +1
#    if y == 199:
#        print(f'Batch complete: {x}')
#        time.sleep(240)
#        y = 0
```
